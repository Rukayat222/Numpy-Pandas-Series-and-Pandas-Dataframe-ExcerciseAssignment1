##**Two-Dimensional Numpy**

```python
import numpy as np

# Creating the 3x3 NumPy array
patient_data = np.array([[25, 70, 120],
                         [30, 80, 130],
                         [40, 85, 140]])

# Accessing the blood pressure of the second patient (index 1)
second_patient_blood_pressure = patient_data[1, 2]

print("Blood Pressure of the second patient:", second_patient_blood_pressure)

```

```python
import numpy as np

# Creating the 4x4 NumPy array
blood_test_results = np.array([[5.6, 7.1, 6.2, 5.8],
                                [5.3, 6.9, 7.0, 6.5],
                                [5.7, 6.5, 6.8, 6.0],
                                [5.4, 7.0, 6.5, 5.9]])

# Calculating the sum of values across each row
row_sums = blood_test_results.sum(axis=1)

print("Total test results for each patient:")
print(row_sums)
```

```python
import numpy as np

# Creating the 2D NumPy array
blood_sugar_levels = np.array([[80, 85, 90],
                                [95, 92, 88],
                                [85, 83, 82],
                                [100, 105, 110],
                                [75, 78, 80]])

# Finding the average blood sugar level for the 3rd patient (row index 2)
third_patient_average = np.mean(blood_sugar_levels[2])

print("Average blood sugar level for the 3rd patient across all visits:", third_patient_average)
```

```python
import numpy as np

# Creating the BMI matrix (3x2)
bmi_values = np.array([[22.1, 23.5],
                       [27.8, 28.0],
                       [21.2, 20.9]])

# Creating the weight array
weights = np.array([60, 70, 80])

# Reshaping the weight array to align it for element-wise multiplication
weights_reshaped = weights[:, np.newaxis]  # Reshape weights to a column vector

# Performing element-wise multiplication
result = bmi_values * weights_reshaped

print("Result of element-wise multiplication (BMI × Weight):")
print(result)
```

```python
import numpy as np

# Creating the 2x5 NumPy array
temperature_data = np.array([[37.0, 36.5, 37.1, 36.9, 37.0],
                              [36.8, 37.3, 37.0, 36.7, 37.2]])

# Finding the maximum temperature for each patient
max_temperatures = np.max(temperature_data, axis=1)

print("Maximum temperature for each patient:")
print(max_temperatures)
```

## **PANDAS SERIES**

```python
# import pandas
import pandas as pd
```

```python
bp_series = pd.Series([120, 130, 140, 135, 125])

max_bp = bp_series.max()  # Maximum value in the Series
min_bp = bp_series.min()  # Minimum value in the Series

print("Maximum Blood Pressure:", max_bp)
print("Minimum Blood Pressure:", min_bp)
```

```python
cholesterol_series = pd.Series([210, 180, 220, 250, 280, 230])

# Filtering patients with cholesterol levels above 240
high_cholesterol = cholesterol_series[cholesterol_series > 240]  # Returns values greater than 240

# Printing the filtered cholesterol levels
print("Patients with Cholesterol Levels Above 240:\n", high_cholesterol)
```

```python
ages_series = pd.Series([35, 40, 25, 30, 45, 50, 55, 60, 65, 70])

# Calculating the median age of the patients
median_age = ages_series.median()

# Printing the median age
print("Median Age of Patients:", median_age)
```

```python
weight_series = pd.Series([70, 80, 75, 65, 85])

# Adding 5 kg to each patient’s weight
updated_weights = weight_series + 5
# Printing the updated weights
print("Updated Weights:\n", updated_weights)
```

```python
temp_series = pd.Series([37.2, 36.8, 37.5, 38.0, 36.6, 37.1, 36.9, 37.3, 37.4, 37.0])

# Calculating the average temperature of all 10 patients
avg_temp = temp_series.mean()  # Mean is the average of all values in the Series

# Printing the average temperature
print("Average Temperature of Patients:", avg_temp)

# Rounding average temperature to 2 decimal places

avg_temp_rounded = round(avg_temp, 2)

print("Rounded Average Temperature of Patients:", avg_temp_rounded)
```

##**DataFrame**

```python
import pandas as pd
```

```python
#Create a DataFrame with patient data
pdata = { "Name": ["Mark", "Jack", "Spark", "Stark"],
         "Age": [22, 24, 54, 44],
          "Blood Pressure": ["120/80", "140/90", "110/70", "130/85"],
          "Cholestrol": [200, 210, 220, 280 ]
          }
#Create the dataframe
df = pd.DataFrame(pdata)

# Select all patients with cholesterol greater than 220
hc_patient = df[df["Cholestrol"] > 220]
print(hc_patient)
```

```python
# Creating a DataFrame with patient information
pidata = {
    "Name": ["Alice", "Bob", "Charlie", "Diana"],
    "Age": [25, 30, 35, 40],
    "Height": [1.65, 1.75, 1.80, 1.60],  # Heights in meters
    "Weight": [55, 68, 77, 60],         # Weights in kilograms
    "Blood Pressure": ["120/80", "130/85", "140/90", "125/82"]
}

pid_df = pd.DataFrame(pidata)

# Calculating BMI and adding it as a new column
pid_df["BMI"] = pid_df["Weight"] / (pid_df["Height"] ** 2)

# Displaying the updated DataFrame
pid_df
```

```python
# Creating a DataFrame with blood test results
blood_test_data = {
    "Test ID": [1, 2, 3, 4],
    "Patient Name": ["John", "Jane", "Tom", "Alice"],
    "Result": [5.6, 6.1, 5.8, 6.3]
}

blood_test_df = pd.DataFrame(blood_test_data)

# Filtering patients with a test result greater than 6.0
filtered_patients = blood_test_df[blood_test_df["Result"] > 6.0]

# Displaying the filtered DataFrame
filtered_patients
```

```python
# Creating a DataFrame with patient data
data = {
    "Age": [25, 34, 45, 29, 61],
    "Weight": [68, 75, 80, 70, 85],  # in kg
    "Height": [165, 170, 175, 160, 180]  # in cm
}

df = pd.DataFrame(data)

# Calculating the mean of each column
average_values = df.mean()

print(average_values)
```

```python
# Creating a DataFrame with treatment details
data = {
    'Patient Name': ['John', 'Jane', 'Tom', 'Alice'],
    'Treatment Type': ['Surgery', 'Consultation', 'Surgery', 'Consultation'],
    'Treatment Date': ['2022-01-01', '2022-02-01', '2022-03-01', '2022-04-01']
}

treatment_df = pd.DataFrame(data)

# Finding all records where the treatment type is "Surgery"
surgery_records = treatment_df[treatment_df['Treatment Type'] == 'Surgery']

print(surgery_records)
```
